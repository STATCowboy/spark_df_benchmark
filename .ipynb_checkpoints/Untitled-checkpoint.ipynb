```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('results.csv')
df=df.drop(df.columns[0], axis=1)
df.head()
```

```python
df['row_count']=df['element_count']/df['column_count']
df.sort_values(by=['row_count', 'column_count'])
```

```python
df[['row_count', 'pandas_df_compute']].plot(kind='line', x='row_count', y='pandas_df_compute')
```

```python
df[['column_count', 'row_count', 'pandas_df_compute']].plot(kind='scatter', x='column_count', y='row_count', c='pandas_df_compute')#, s='pandas_df_compute')
```

```python
# df[['column_count', 'row_count', 'pandas_df_compute']].plot(kind='scatter', x='column_count', y='row_count', s='pandas_df_compute')
plt.scatter(x=df['column_count'], y=df['row_count'], s=df['pandas_df_compute']*100)
```
